### Embedding Layer

```python
def embed():
    inp = Input(shape=(100,100,3))
    
    cnv1 = Conv2D(64, (10,10), activation='relu')(inp)
    mxp1 = MaxPooling2D(64, (2,2), padding='same')(cnv1)
    
    cnv2 = Conv2D(128, (7,7), activation='relu')(mxp1)
    mxp2 = MaxPooling2D(64, (2,2), padding='same')(cnv2)
    
    cnv3 = Conv2D(128, (4,4), activation='relu')(mxp2)
    mxp3 = MaxPooling2D(64, (2,2), padding='same')(cnv3)
    
    cnv4 = Conv2D(256, (4,4), activation='relu')(mxp3)
    f1 = Flatten()(cnv4)
    d1 = Dense(4096, activation='sigmoid')(f1)
    
    return Model(inputs=[inp], outputs=[d1], name='embedding')
```

```python
embedding = embed()
```

### Distance Layer

```python
# Siamese distance class
class L1Dist(Layer):
    
    # init method for inheritance
    def __init__(self, **kwargs):
        super().__init__()
    
    # difference calculation of input and validation images
    def diff(self, input_embedding, val_embedding):
        return tf.math.abs(input_embedding - val_embedding)
```

```python
L1 = L1Dist()
```

### Siamese Neural Network

```python
def siamese_model():
    
    # anchor image input
    input_image = Input(name='input_image', shape=(100,100,3))
    
    # validation image input
    val_image = Input(name='val_image', shape=(100,100,3))
    
    # combining siamese distance components
    siamese_layer = L1Dist()
#     siamese_layer.name = 'distance'
    distances = siamese_layer(embedding(input_image), embedding(val_image))
    
    # final classification layer
    classifier = Dense(1, activation='sigmoid')(distances)
    
    return Model(inputs=[input_image, val_image], outputs=classifier, name='SiameseNetwork')
    
```

```python
# training step function
# training one batch includes -
# 1. making a prediction
# 2. calculating loss
# 3. deriving gradients
# 4. calculating new weights and applying them
```

```python
# @tf.function will compile whatever is gonna happen in the function below
@tf.function
def train_step(batch):
    
    # tf.GradientTape records or captures gradients from our neural network model
    with tf.GradientTape() as tape:
        # grabbing anchor image and positive or negative image
        x = batch[:2]
        # grabbing the label
        y = batch[2]
        
        # forward pass
        ypred = siamesemodel(x, training=True)
        
        
        # calculating loss
        loss = binary_cross(y, ypred)
    print(loss)
        
    # calculating gradients
    grad = tape.gradient(loss, siamesemodel.trainable_variables)  
    
    # calculating updated weights and applying them to siamese model
    opt.apply_gradients(zip(grad, siamesemodel.trainable_variables))
        
    return loss
```

### building training loop

```python
def train(data, epochs):
    
    # looping through each epoch
    for epoch in range(1, epochs+1):
        print('\n Epoch{}/{}'.format(epoch, epochs))
        progbar = tf.keras.utils.Progbar(len(data))
        
        # looping through each batch of data
        for idx, batch in enumerate(data):
            
            # running train step function
            train_step(batch)
            progbar.update(idx+1)
```

### Verification function

```python
def verify(model, detection_threshold, verification_threshold):
    results = []
    
    for image in os.listdir(os.path.join('app_data','verification_images')):
        input_image = preprocessing(os.path.join('app_data/input_images/input_image.jpg'))
        val_image = preprocessing(os.path.join('app_data', 'verification_images', image))
        
        # predictions
        result = model.predict(list(np.expand_dims([input_image, val_image], axis=1)))
        results.append(result)
        
    # Detection threshold - above this threshold a prediction will be considered positive
    detection = np.sum(np.array(results) > detection_threshold)
    
    # verification threshold - ratio of positive predictions to total positive samples
    verification = detection / len(os.listdir(os.path.join('app_data', 'verification_images')))
    verified = verification > verification_threshold
    
    
    return results, verified
        
```
